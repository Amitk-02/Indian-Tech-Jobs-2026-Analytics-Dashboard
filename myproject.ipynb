```python
comp = df['company_name'] 
company_name = []
for i in comp:
      company_name.append(i)
print(company_name)
```

```python
company_dict = {}
for i in company_name:
    if i in company_dict:
        company_dict[i] += 1
    else:
        company_dict[i] = 1
print(company_dict)
```

```python

#sort company_dict in assending order
sorted_company = dict(sorted(company_dict.items(), key = lambda x :x[1], reverse = True))
print(sorted_company)
```

```python
#top 10 companies names 
comptop_10=[]
for i in sorted_company:
    if len(comptop_10)<10:
        comptop_10.append(i)

print(comptop_10)
print(len(comptop_10))

# top 10 compines counts
counttop_10=[]
for i in sorted_company:
    if len(counttop_10)<10:
        counttop_10.append(sorted_company[i])
print(counttop_10)
print(len(counttop_10))
```

```python
# bar chart 
plt.title('Top 10 Companies with Highest Employee Count')
plt.bar(comptop_10,counttop_10)
plt.xticks(rotation = 90)
plt.xlabel('Company Name')
plt.ylabel('Employee Count')
plt.show()
```

```python
plt.pie(counttop_10, labels = comptop_10, autopct='%1.1f%%',
        shadow=True,
        wedgeprops={"edgecolor":'black', 'width':0.7})
```

```python
# Keep only rows containing 'Yrs'
df = df[df['experience_raw'].str.contains('Yrs', na=False)]

# Remove ' Yrs'
exp = df['experience_raw'].str.replace(' Yrs', '', regex=False)

# Split at '-'
exp = exp.str.split('-', expand=True)

# Convert to numeric
df['min_exp'] = pd.to_numeric(exp[0], errors='coerce')
df['max_exp'] = pd.to_numeric(exp[1], errors='coerce')

# Calculate average experience
df['avg_experience'] = (df['min_exp'] + df['max_exp']) / 2

# View result
df[['experience_raw', 'min_exp', 'max_exp', 'avg_experience']].head()
```

```python


exp_count = (
    df['experience_raw']
      .value_counts()
      .head(10)
      .reset_index()
)

exp_count.columns = ['Experience', 'Jobs']

fig = px.bar(
    exp_count,
    x='Experience',
    y='Jobs',
    color='Jobs',
    text='Jobs',
    title='Top 10 Most Common Experience Requirements'
)

fig.update_traces(textposition='outside')
fig.update_layout(template='plotly_white')

fig.show()
```
